```python
#input user name to return basic greeting
user_name = input("What is your name? ")
print("Hello, " + user_name + "! Welcome to Walesia's magical INST126 Calculator!")

import math  #import package to use sum() and math.prod() functions

# The functions below are defined to run after the user selects the corresponding operation
def addition(listofnums): 
    return sum(listofnums) #using math package

def subtraction(n1, n2): 
    return n1 - n2

def multiplication(listofnums): 
    return math.prod(listofnums) #using math package

def division(n1, n2): 
    return n1 / n2

def exponentation(n1, n2): 
    return n1**n2

def square_root(n1): 
    return math.sqrt(abs(n1)) #use the abs() to turn an accidental negative input into a positive number for function to work

#The following while loop is an indefinite iteration which will re-run calculator after an operation is complete 
while True: 
    
    #user prompt for calculator selection
    user_selection = input(''' 
First, please choose an operation: 
1. Addition
2. Subtraction
3. Multiplication
4. Division
5. Exponentation
6. Square Root
7. Quit

''') 

    if user_selection in ('2', '4', '5', '6'): #if user selection is 2, 4, 5, 6 - note special directions
        print("Next, follow the directions in the prompt(s) below: ")
        print(" ")
    
    # If statements below check to see if user's input is one of the 6 valid options
    # If... elif... statements execute options and call our operator functions
    
    #Conditional for addition 
    if user_selection == '1': 
        listofnums = []
        n = int(input("How many numbers do you want to add?: ")) #Bonus: User can choose how many numbers to add. 
        #Assuming a valid integer input from user
        print(" ")
            
        for i in range(0, n): #loop prompts user for however many numbers they requested to add
            print("Enter number #", i+1, ":")
            user_list = int(input())
            listofnums.append(user_list)
            
        print(" ")
        print("These numbers added together equal ", addition(listofnums))
        print(" ")
        print("Thanks for trying the addition function! Let's do another one! ")
     
    #Conditional for subtraction 
    elif user_selection == '2':
        firstnumber = float(input("Enter your first number: "))
        secondnumber = float(input("Enter your second number: "))
        print(firstnumber, "-", secondnumber, "=", subtraction(firstnumber, secondnumber))
        
    #Conditional for multiplication
    elif user_selection == '3':
        listofnums = [] 
        n = int(input("How many numbers do you want to multiply? : ")) #Again, assuming a valid integer input here
        print("\n")
        
        for i in range(0, n): 
            print("Enter number #", i+1, ":") #iteration asks for input for however many numbers requested
            user_list = int(input())
            listofnums.append(user_list)
        print(" ")
        print("The product of these numbers is ", multiplication(listofnums))
        print(" ")
        print("Thank you for trying the multiplication function! Let's do another one! ")
    
    #Conditional for division
    elif user_selection == '4': 
        firstnumber = float(input("Enter the dividend, aka the number to be divided: "))
        secondnumber = float(input("Enter the divisor (not zero), aka the number to divide by: "))
        print(firstnumber, "/", secondnumber, "=", division(firstnumber, secondnumber))
        print(" ")
        print("Thank you for trying the division function! Let's try another one! ")
    
    #Bonus: Conditional for exponentation
    elif user_selection =='5':
        firstnumber = float(input("Enter the base number: "))
        secondnumber = float(input("Enter the exponent number: "))
        print(firstnumber, "^", secondnumber, "=", exponentation(firstnumber, secondnumber))
        print(" ")
        print("Thank you for trying the exponential function! Let's try another one! ")
     
    #Bonus: Conditional for finding the square root 
    elif user_selection =='6':
        firstnumber = float(input("Enter a number to find the square root of: "))
        print("The âˆš", abs(firstnumber), " = ", square_root(firstnumber))
        print(" ")
        print("Thank you for trying the square root function! Let's try another one! ")
    
    #If user chooses to quit the program
    elif user_selection == '7':
        print("Okay, bye then! ")
        break
       
    #If user did not choose 1-6, return invalid input statement and our loop restarts until an acceptable choice is made.
    else:
        print("Sorry, that's not a valid input. Please choose again. ")
        
```
